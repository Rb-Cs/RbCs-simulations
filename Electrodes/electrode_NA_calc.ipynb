```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Assuming a square spacing of the rods
rod_diameters = np.linspace(2,4,1000)
rod_spacing = np.linspace(9,28,1000)
diameters, spacing = np.meshgrid(rod_diameters, rod_spacing)
    
def findNA(r, L):
    return np.sin( np.arccos(r/(np.sqrt(2)*L)) - (np.pi/4))

Z = findNA(diameters, spacing)
plt.imshow(Z, 
           origin='lower',
           extent=[diameters.min(), diameters.max(), spacing.min(), spacing.max()],
           aspect='auto')
plt.colorbar()
plt.contour(diameters, spacing, Z, levels=[0.6], colors='red', linewidths=2, label='0.6 NA')
plt.contour(diameters, spacing, Z, levels=[0.65], colors='orange', linewidths=2, label='0.6 NA')
plt.plot(2.39, 11.324, marker='*', color='green') # Same groove grabbers as Yb
plt.plot(3.175, 2*7.62, marker='*', color='blue') # These groove grabbers https://www.kimballphysics.com/product/mcf275-grvgrb-cb04/
plt.xlabel("Diameter of rods (mm)")
plt.ylabel("Spacing b/w rods (mm)")
plt.title("NA for a square arrangement of electrode rods")
```
