```python
import numpy as np
from scipy.constants import c, k as k_B, physical_constants
```

```python
# function to calculate the derivative of C
def dC_dz(alpha, wavelength):
    # alpha is the NA of the objective, wavelength is the wavelength of the trapping light
    # return in units / um^6
    return 2.6*alpha * np.sin(alpha) / wavelength / 1e6  # convert from 1/m to 1/um
```

First using the parameters in the paper.

```python
# calculation of the fictitious B field gradient caused by polarization of tweezers referencing this paper: https://journals.aps.org/prl/abstract/10.1103/PhysRevLett.110.133001

# Paper parameters
lambda_trap = 815e-9          # m
U0_mK = 0.82                  # mK
NA  = 0.43                 # Numerical aperture of the objective
dCxdY = dC_dz(NA, lambda_trap)                  # 1/um
print(f"dC_x/dy: {dCxdY:.6f} 1/um")

# 87Rb vacuum wavelengths, Steck Rb87 D-line data
lambda_D1 = 794.978851156e-9  # m, 5S1/2 -> 5P1/2
lambda_D2 = 780.241209686e-9  # m, 5S1/2 -> 5P3/2

mu_B = physical_constants["Bohr magneton"][0]

# Frequencies
nu_trap = c / lambda_trap
nu_D1 = c / lambda_D1
nu_D2 = c / lambda_D2

# Detunings in Hz. Angular-frequency detunings give the same ratio.
delta_1 = nu_trap - nu_D1
delta_2 = nu_trap - nu_D2

vector_factor = (delta_2 - delta_1) / (delta_2 + 2 * delta_1)

# Convert trap depth U0 = k_B T to equivalent B field: U0 / mu_B
U0_J = k_B * U0_mK * 1e-3
B0_equiv_T = U0_J / mu_B
B0_equiv_G = B0_equiv_T * 1e4

# Effective fictitious magnetic-field gradient
B_gradient_G_per_um = B0_equiv_G * vector_factor * dCxdY

print("87Rb vector-light-shift fictitious B-field gradient")
print("--------------------------------------------------")
print(f"Trap wavelength: {lambda_trap*1e9:.3f} nm")
print(f"D1 wavelength:   {lambda_D1*1e9:.12f} nm")
print(f"D2 wavelength:   {lambda_D2*1e9:.12f} nm")
print()
print(f"delta_1 / 2pi equivalent: {delta_1/1e12:.6f} THz")
print(f"delta_2 / 2pi equivalent: {delta_2/1e12:.6f} THz")
print(f"vector factor:            {vector_factor:.6f}")
print()
print(f"U0 / mu_B:                {B0_equiv_G:.6f} G")
print(f"dC_x/dy:                  {dCxdY:.6f} 1/um")
print(f"B'_x gradient:            {B_gradient_G_per_um:.6f} G/um")
```
