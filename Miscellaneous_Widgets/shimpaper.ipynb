Small notebook to calculate the optimal combination of shim papers

```python
import itertools
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

variables = {
    "red": 0.08,
    "green": 0.11,
    "blue": 0.15,
    "black": 0.31,
    "pink": 0.45,
    "yellow": 0.53,
    "white": 0.70,
    "orange": 0.85,
}

# Text colors for visibility
display_colors = {
    "red": "red",
    "green": "green",
    "blue": "blue",
    "black": "black",
    "pink": "hotpink",
    "yellow": "goldenrod",
    "white": "gray",      # white text is invisible on white background
    "orange": "orange",
}

max_terms = 5
names = list(variables)

# Precompute combinations
all_combos = []

for r in range(1, max_terms + 1):
    for combo in itertools.combinations_with_replacement(names, r):
        total = sum(variables[name] for name in combo)
        all_combos.append((combo, total))

def closest_combo(target):
    return min(all_combos, key=lambda item: abs(item[1] - target))

slider = widgets.FloatSlider(
    value=1.0,
    min=0.0,
    max=5 * max(variables.values()),
    step=0.01,
    description="Target",
    continuous_update=False,
)

box = widgets.BoundedFloatText(
    value=1.0,
    min=0.0,
    max=5 * max(variables.values()),
    step=0.01,
    description="Value",
)

output = widgets.Output()

updating = False

def colored_expression(combo):
    parts = []

    for name in combo:
        color = display_colors[name]
        value = variables[name]

        html = (
            f'<span style="color:{color}; '
            f'font-weight:bold; font-size:16px;">'
            f'{name}({value})'
            f'</span>'
        )

        parts.append(html)

    return " + ".join(parts)

def update(target):
    combo, total = closest_combo(target)

    expr_html = colored_expression(combo)

    with output:
        clear_output(wait=True)

        display(HTML(f"""
        <div style="font-family: monospace; font-size:15px;">
            <p><b>Target:</b> {target:.4f}</p>
            <p><b>Closest sum:</b> {total:.4f}</p>
            <p><b>Difference:</b> {abs(total-target):.4f}</p>
            <p><b>Variables:</b><br>{expr_html}</p>
        </div>
        """))

def slider_changed(change):
    global updating

    if updating:
        return

    updating = True
    box.value = change["new"]
    update(change["new"])
    updating = False

def box_changed(change):
    global updating

    if updating:
        return

    updating = True
    slider.value = change["new"]
    update(change["new"])
    updating = False

slider.observe(slider_changed, names="value")
box.observe(box_changed, names="value")

display(widgets.VBox([slider, box, output]))

update(1.0)
```

```python
def color_sum(*colors):
    """
    Example:
        color_sum("red", "blue", "orange")
    """

    total = 0.0

    for color in colors:
        if color not in variables:
            raise ValueError(f"Unknown color: {color}")

        total += variables[color]

    return round(total, 3)


color_sum("orange", "pink", "green")
```
